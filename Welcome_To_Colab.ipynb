```
from itertools import combinations

nums = [-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4]

def find_zero_sum_subsets_fixed_size(nums, size=5):
    result = []
    for subset in combinations(nums, size):
        if sum(subset) == 0:
            result.append(subset)
    return result

# Call for problem 1
result_5 = find_zero_sum_subsets_fixed_size(nums, size=5)
print(f"Subsets of size 5 with sum 0:\nTotal: {len(result_5)}")
for subset in result_5:
    print(subset)
```

```
def find_zero_sum_subsets_range_size(nums, min_size=3, max_size=6):
    result = []
    for size in range(min_size, max_size + 1):
        for subset in combinations(nums, size):
            if sum(subset) == 0:
                result.append(subset)
    return result

# Call for problem 2
result_3_6 = find_zero_sum_subsets_range_size(nums, 3, 6)
print(f"Subsets of size 3 to 6 with sum 0:\nTotal: {len(result_3_6)}")
for subset in result_3_6:
    print(subset)
```
